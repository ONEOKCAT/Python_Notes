**# 绘图基础**

**1. 简单图形绘制**

```python
from matplotlib import pyplot as plt
import numpy as np
import math #needed for definition of pi
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show()
```

**2. 关于ax=fig.add_axes([0,0,1,1])**

* add_axes()方法需要一个由4个元素组成的list对象，分别对应图形的左、底、宽、高。每个数字必须在0和1之间。
* 例如，改动为 ax= fig.add_axes([0.8,0,1,1])，改变第一个参数为0.8，图像从左往右移动了80%；改动为ax = fig.add_axes([0,0.8,1,1])，改变第二个参数为0.8，图像从下往上移动了80%。

```python
from matplotlib import pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)

fig = plt.figure()  # 提供一个空的画布
ax = fig.add_axes([0,0,1,1])  # 为图形添加轴 

ax.plot(x,y)
plt.show()
```

* 不同效果查看

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)

fig = plt.figure()  # 提供一个空的画布
ax = fig.add_axes([0,0,0.5,1])  # 为图形添加轴 

ax.plot(x,y)
plt.show()
```

**3. Figure类和Axes类**

**Figure类**

* matplotlib.figure模块包含了Figure类。它是所有绘图元素的顶层容器。Figure对象是通过调用pyplot模块中的fig()函数来实例化的，如下所示。(简单来说，Figure就是给我们画图提供了一个画布，这个画布的参数我们可以通过Figure类进行调整和修改，画布弄好之后才方便在上面作图)

* fig = plt.figure()，包含如下参数:
  1. figsize	(宽，高)，是个元组，表示Figure的大小
  2. dpi	每英寸的点数，也就是分辨率
  3. facecolor	画布的颜色
  4. edgecolor	画布的边缘颜色
  5. linewidth	边缘的线宽

```python
from matplotlib import pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)

fig = plt.figure(figsize=(8,4),dpi=144,facecolor='g')
ax = fig.add_axes([0,0,0.3,0.3])

ax.plot(x,y)
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')
plt.show()
```

**Axes类**

* Axes对象是图像中具有数据空间的区域。一个给定的图可以包含许多Axes，但一个给定的Axes对象只能在一个图中。通过调用add_axes()方法将轴对象添加到图形中。

* 函数1 Legend: axes类的legend()方法为绘图添加了一个图例。它需要三个参数：
  
  ax.legend(handles, labels, loc)

* Location string和Location code对应如下

  best-0, upper right-1, upper left-2,

  lower left-3, lower right-4, Right-5,

  Center left-6, Center right-7, lower center-8,

  upper center-9, Center-10

* 函数2 axes.plot(): 这是axes类的基本方法，它将一个数组的值与另一个数组的值以线或标记的形式绘制出来。plot()方法可以有一个可选的格式字符串参数来指定线条和标记的颜色、样式和大小。

* 颜色: 

  b-blue, g-green, r-red, w-white, c-cyan, m-magenta, y-yellow, k-black

* 样式(点):

  . - Point marker, o - Circle marker, x - X marker, D - Damond marker,

  H - Hexagon marker, s - Square marker, + - PLus marker

* 线型:

  '-' - Soild line, '--' - Dashed line, '-.' Dash-dot line, 
  ':' - Dooted line 


```python
# 例1
import numpy as np

x = np.arange(20)

fig = plt.figure()
ax = plt.subplot(111)

for i in range(4):
    ax.plot(x, i * x, label='$y = %ix$' % i)

ax.legend(loc=0)

plt.show()
```

```python
# 例2

y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

l1 = ax.plot(x1,y,'ys-') # solid line with yellow colour and square marker
l2 = ax.plot(x2,y,'go--') # dash line with green colour and circle marker

ax.legend(labels = ('tv', 'Smartphone'), loc = 'lower right') # legend placed at lower right
ax.set_title("Advertisement effect on sales")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
plt.show()
```

**4. 绘图常用方式分析**

* pyplot绘图: 第一种是最常见的，即通过调用pyplot的api去进行绘图。

* matplotlib.pyplot (opens new window)是一组命令样式函数，使Matplotlib的工作方式类似于MATLAB。每个pylot函数对图形进行一些更改：例如，创建图形、在图形中创建绘图区域、在绘图区域中绘制一些线、使用标签装饰绘图等。pyplot (opens new window)主要用于交互式绘图和编程绘图生成简单图例。

* 面向对象的API: Matplotlib的核心是面向对象的。如果需要对 plots 进行更多控制和自定义，我们建议直接使用对象。在许多情况下，将使用 pyplot.subplots创建一个图形和一个或多个轴，然后处理这些对象。

* 这种方式是比较推荐使用的，较pyplot的方式可以进行更细致的精调。
即fig, ax = plt.subplots()，指定figure和axes，然后对axes单独操作。
