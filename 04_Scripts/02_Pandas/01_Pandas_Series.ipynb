**# Pandas_Series**

**Series基本概念**

* Series， 带标签的一维数组

```python
import pandas as pd

a = pd.Series([1, 2, 31, 12, 4, 3])
a
```

```python
type(a)
```

* 可以指定索引，长度需要保持一致

```python
b = pd.Series([1, 2, 31, 12, 4, 3], index = list('abcdef'))
b
```

* 修改数据类型(和numpy一样)

```python
b.astype(float)
```

* 对于字典类型的转换

```python
dict = {'name': 'Bob', 'age': 24, 'tel': 10086}

pd.Series(dict)
```

* 对于已有数据之外的索引会显示NaN

```python
import string

a = {string.ascii_uppercase[i]: i for i in range(10)}
a
```

```python
pd.Series(a, index = list(string.ascii_uppercase[5:15]))
```

**Series的切片和索引**

```python
a = pd.Series(a, index = list(string.ascii_uppercase[5:15]))
a
```

```python
a[1], a['I']
```

```python
a[[2,3]]
```

```python
a[2:5]
```

```python
a[0:10:2]
```

```python
a[a>4]
```

**Series的迭代和值**

```python
dict = {'name': 'Bob', 'age': 24, 'tel': 10086}

a = pd.Series(dict)
a
```

```python
a.index
```

```python
for i in a.index:
   print(i)
```

```python
len(a.index)
```
